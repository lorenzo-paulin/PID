```python
%%manim -qh Gaussiana
# config.media: tamanho que o video aparece aqui no colab
config.media_width= "80%"
from manim.mobject.geometry.tips import ArrowTriangleTip

class Gaussiana(Scene):

#funo pra gerar as flechas
  def angulo_flechas(self, n):
    theta_rad = (2*np.pi)/n
    return theta_rad

  def coordenadas_flecha (self, x_01, x_11, y_01, y_11, n):

    theta = self.angulo_flechas(n)

    rot = np.array([[np.cos(theta), - np.sin(theta)], [np.sin(theta), np.cos(theta)]])

    new_coord_init = np.dot(rot, np.array([x_01, y_01]))

    new_coord_final = np.dot(rot, np.array([x_11, y_11]))

    return new_coord_init, new_coord_final


  def construct(self):

    #pontos de inicio e fim

    #flechas
    x_01 = 0
    y_01 = 0.5

    x_11 = 0
    y_11 = 2.5

    #linhas
    xl_01 = 0
    yl_01 = 0.5

    xl_11 = 0
    yl_11 = 8


    arrowpositiva = Arrow([x_01, y_01, 0], [x_11, y_11,0], buff=0)

    #carga
    circulo_positivo= Circle(radius=0.4, fill_opacity=1, color=PURE_RED)
    texto_positivo= Text('+', color=BLACK)

    nro_flechas = 16


    #flechas da carga
    Grupoflechaspositivas = VGroup()
    for i in range (nro_flechas):

      new_coord_init, new_coord_final = self.coordenadas_flecha(x_01, x_11, y_01, y_11, nro_flechas)

      Grupoflechaspositivas += Arrow([new_coord_init[0], new_coord_init[1], 0],
                                     [new_coord_final[0], new_coord_final[1],0],
                                     buff=0,
                                     stroke_width=1,
                                     max_tip_length_to_length_ratio=0.075)

      x_01, x_11, y_01, y_11 = new_coord_init[0], new_coord_final[0], new_coord_init[1], new_coord_final[1]
    grupo_positivo = Group(circulo_positivo, texto_positivo)

    #linhas estendidas
    Glpositivas= VGroup()
    for i in range (nro_flechas):

      new_coord_init, new_coord_final = self.coordenadas_flecha(xl_01, xl_11, yl_01, yl_11, nro_flechas)

      Glpositivas += Line(
          [new_coord_init[0], new_coord_init[1], 0],
          [new_coord_final[0], new_coord_final[1],0],
          buff=0)

      xl_01, xl_11, yl_01, yl_11 = new_coord_init[0], new_coord_final[0], new_coord_init[1], new_coord_final[1]

    #risco vet da

    curva = ParametricFunction(
            lambda u: np.array((
                1.5* np.cos(u),
                1.5* np.sin(u),
                0
            )), color= PURE_GREEN, t_range = np.array([1,1.3]))

    riscoazul= curva.move_to(LEFT*3.4 + 0.685*UP )

    #flechas da
    xv_01 = 0
    yv_01 = 1.5

    xv_11 = 0
    yv_11 = 2

    Grupoda = VGroup()
    for i in range (nro_flechas):

      new_coord_init, new_coord_final = self.coordenadas_flecha(xv_01, xv_11, yv_01, yv_11, nro_flechas)

      Grupoda += Arrow([new_coord_init[0], new_coord_init[1], 0],
                                     [new_coord_final[0], new_coord_final[1],0],
                                     buff=0,
                                     stroke_width=10,
                                     max_tip_length_to_length_ratio=0.3,
                                     color= PURE_GREEN)

      xv_01, xv_11, yv_01, yv_11 = new_coord_init[0], new_coord_final[0], new_coord_init[1], new_coord_final[1]

    vetda= Grupoda[14].move_to(LEFT*3.32 + 0.95*UP)



    #equaes
    X= 2.5*RIGHT
    T= 60
    Formula = MathTex(r"\oint _S \vec{E} \cdot", r" d\vec{A}", r"=", r"{q_{int}", r"\over\epsilon_0}",font_size=T).move_to(X)
    Formula2=MathTex(r"\oint _S |d\vec{A}| |\vec{E}| cos(\theta)=\frac{q_{int}}{\epsilon_0}", font_size=T ).move_to(X)
    Formula2_1=MathTex(r"\oint _S |d\vec{A}| |\vec{E}| cos(0)=\frac{q_{int}}{\epsilon_0}",font_size=T ).move_to(X)
    Formula3=MathTex(r"\oint _S |d\vec{A}| |\vec{E}| 1 =\frac{q_{int}}{\epsilon_0}", font_size=T ).move_to(X)
    Formula4=MathTex(r"\oint", r"E \, dA", r"= \frac{+q}{\epsilon_0}", font_size=T ).move_to(X)
    Formula5=MathTex(r" E \oint  dA", r"= \frac{q}{\epsilon_0}", font_size=T ).move_to(X)
    Formula6=MathTex(r" E (4 \pi r^{2})", r"= \frac{q}{\epsilon_0}", font_size=T ).move_to(X)
    Formula7=MathTex(r" E = \frac{1}{4 \pi r^{2}}\frac{q}{\epsilon_0}", font_size=T).move_to(X)
    Formula8=MathTex(r" \vec{E}", r"= \frac{1}{4 \pi r^{2}}\frac{q}{\epsilon_0}", r"\hat{r}", font_size=T).move_to(X)
    quadradofinal = SurroundingRectangle(Formula8, buff = .1, color=PURE_RED).move_to(X)

    #itens amarelos para explicar a equao (em ordem)
    b1 = Brace(Formula6[0],color=YELLOW).move_to(2.2*RIGHT+ DOWN)
    E_= MathTex(r"\vec{E}= E_{r} \cdot \hat{r}", font_size= 45, color=YELLOW).move_to(2.3*RIGHT+1.5*DOWN)
    da_= MathTex(r"d\vec{A}= dA \cdot \hat{r}", font_size= 45, color=YELLOW).move_to(2.3*RIGHT+2*DOWN)
    theta= MathTex(r"\vec{E} \parallel d\vec{A} \Rightarrow \theta=0", color=YELLOW).move_to(2*RIGHT+1.7*DOWN)
    qint=MathTex(r"q_{int}=+q", color=YELLOW).move_to(2.3*RIGHT+1.7*DOWN)
    leic= Text(r"Da Lei de Coulomb:", font_size= 25, color=YELLOW).move_to(2.15*RIGHT+1.5*DOWN)
    leic2= Text(r"o campo eltrico  constante ", font_size= 25, color=YELLOW).next_to(leic, DOWN)
    A=MathTex(r" A = 4 \pi r^{2}", font_size= 40, color=YELLOW).move_to(2*RIGHT+1.5*DOWN)

    #itens junto a carga
    circulo = Circle(radius=1.5, color=PURE_RED).move_to(LEFT*4 + 0.7*DOWN)
    texto = Text('Lei de Gauss', color=PURE_RED).move_to(3*UP)
    textDA = MathTex(r'd\vec{A}', color=  PURE_GREEN).move_to(2*RIGHT)
    E= MathTex(r"\vec{E}").move_to(4*LEFT+2.5*UP)
    textQ= MathTex(r'+q', color=PURE_RED).move_to(0.2*UP+2.85*RIGHT)


    #grupo das animaes das flechas e linhas estendidas
    g1 = Group(grupo_positivo,Grupoflechaspositivas)

    #obs:
    #run_time=x  o tempo que a animao dura
    #wait(x)  quantos segundos ele espera para gerar a prox animao

    self.play(FadeIn(texto,run_time=5))
    self.add(grupo_positivo)
    self.play(FadeIn(Glpositivas, Grupoflechaspositivas, run_time = 2))
    self.play(FadeOut(Glpositivas))
    self.play(g1.animate.shift(LEFT*4 + 0.7*DOWN))
    self.play(GrowFromCenter(circulo, run_time = 1.5))
    self.play(Write(Formula))
    self.play(FadeIn(E))
    self.play(Indicate(Formula[1], color=PURE_GREEN))
    self.play(FadeIn(curva),GrowArrow(vetda),textDA.animate.shift(LEFT*4.4+ UP*2.1))
    self.play(Indicate(Formula[3], color=PURE_RED ))
    self.play(textQ.animate.shift(5.75*LEFT + 0.63*DOWN))
    self.add(b1)
    self.play(Write(E_))
    self.play(Write(da_))
    self.wait()
    self.play(TransformMatchingShapes(Formula, Formula2, run_time = 1.5))
    self.play(FadeOut(E_,da_))
    self.wait()
    self.play(Write(theta))
    self.play(TransformMatchingShapes(Formula2, Formula2_1,run_time = 1.5))
    self.play(TransformMatchingShapes(Formula2_1, Formula3, run_time = 1.5))
    self.play(FadeOut( theta, b1 ))
    self.play(FadeIn(b1, qint))
    self.wait()
    self.play(TransformMatchingShapes(Formula3, Formula4, run_time = 1.5))
    self.play(FadeOut(qint))
    self.play(Write(leic))
    self.play(Write(leic2))
    self.play(TransformMatchingShapes(Formula4, Formula5, run_time = 1.5), FadeOut(b1,leic, leic2))
    self.play(FadeIn(b1),Write(A))
    self.play(TransformMatchingShapes(Formula5, Formula6, run_time = 1.5))
    self.play(FadeOut(A, b1))
    self.play(TransformMatchingShapes(Formula6, Formula7))
    self.play(TransformMatchingShapes(Formula7, Formula8, run_time=1.5))
    self.play(Indicate(Formula8[0], color=PURE_RED), Indicate(Formula8[2], color=PURE_RED))
    self.play(Create(quadradofinal))
    self.wait(10)
```
