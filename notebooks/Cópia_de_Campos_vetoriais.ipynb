```python
%%manim -ql campocerto

class campocerto(Scene):
    def construct(self):
        import numpy as np

        rplano= Rectangle(width=0.1, height=4.5, color=YELLOW,fill_opacity=1).move_to(point_or_mobject=[-4,0,0])
        da1= Arrow(start=np.array([-4.20, 0, 0.]), end=np.array([-1.5, 0, 0]),color=BLUE)

        planofinal= rplano.rotate(about_point= [-4,0,0] ,axis=np.array([0., 0., 1.]), angle=PI/2 )

        #grupo3= grupo1.copy().rotate(about_point=[-4,0,0],axis=np.array([0., 0., 1.]), angle=PI/2 )

        titulo= Paragraph(r"Fluxo de um Campo Vetorial", r"através de uma superfície", font_size=40, alignment='center', color=YELLOW).move_to(point_or_mobject=[0,0,0])



#LINHAS DE CAMPO

        f1_= Arrow(start=np.array([-15., 3, 0.]), end=np.array([12, 3, 0]),color=GRAY_D)
        f2_= Arrow(start=np.array([-15., 2,0]), end=np.array([12, 2, 0]),color=GRAY_D)
        f3_= Arrow(start=np.array([-15., 1, 0.]), end=np.array([12, 1, 0.]),color=GRAY_D)
        f4_= Arrow(start=np.array([-15., 0, 0]), end=np.array([12, 0, 0]),color=GRAY_B)
        f5_= Arrow(start=np.array([-15., -1, 0]), end=np.array([12,-1, 0]),color=GRAY_D)
        f6_= Arrow(start=np.array([-15., -2,0]), end=np.array([12, -2, 0]),color=GRAY_D)
        f7_= Arrow(start=np.array([-15., -3, 0.]), end=np.array([12, -3, 0.]),color=GRAY_D)
        GRUPOLINHASinicio= Group(f1_,f2_,f3_,f5_, f6_,f7_)

        f1= Arrow(start=np.array([-7., 3, 0.]), end=np.array([-0.5, 3, 0]),color=GRAY_D)
        f2= Arrow(start=np.array([-7., 2,0]), end=np.array([-0.5, 2, 0]),color=GRAY_D)
        f3= Arrow(start=np.array([-7., 1, 0.]), end=np.array([-0.5, 1, 0.]),color=GRAY_D)
        f4= Arrow(start=np.array([-7., 0, 0]), end=np.array([-0.5, 0, 0]),color=GRAY_B)
        f5= Arrow(start=np.array([-7., -1, 0]), end=np.array([-0.5,-1, 0]),color=GRAY_D)
        f6= Arrow(start=np.array([-7., -2,0]), end=np.array([-0.5, -2, 0]),color=GRAY_D)
        f7= Arrow(start=np.array([-7., -3, 0.]), end=np.array([-0.5, -3, 0.]),color=GRAY_D)
        GRUPOLINHAS= Group(f1,f2,f3,f4,f5, f6,f7)

### PARA 0 GRAUS
        plano__= rplano1= Rectangle(width=0.1, height=4.5, color=YELLOW,fill_opacity=1).move_to(point_or_mobject=[-4,0,0])
        planocopia= plano__.copy()

        da0= Arrow(start=np.array([-4.20, 0, 0.]), end=np.array([-1.5, 0, 0]),color=BLUE)
        da= Arrow(start=np.array([-4.20, 0, 0.]), end=np.array([-1.5, 0, 0]),color=BLUE)

        textoda0= MathTex(r"d\vec{A}", font_size=45, color=BLUE).next_to(da0,DOWN)
        textoE= MathTex(r"\vec{E}", font_size=45,color=GRAY_B).next_to(f4,0.3*RIGHT+DOWN)

        escrita= MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}| cos(\theta)").move_to(point_or_mobject=[3.5,0,0])
        escrita01= MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}| cos(0)").move_to(point_or_mobject=[3.5,0,0])
        escrita02= MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}| (1)").move_to(point_or_mobject=[3.5,0,0])
        escritamax= Text(r"Fluxo máximo de campo vetorial", font_size=25).move_to(point_or_mobject=[3.5,-2,0])



        GRUPO0= Group(plano__, da0)
        Grupo00= GRUPO0.copy()

## PARA 45 GRAUS
        plano2__= plano__.copy().rotate(about_point=[-4,0,0], axis=np.array([0., 0., 1.]), angle=PI/4 )
        da60= da0.copy().rotate( about_point=[-4,0,0], axis=np.array([0., 0., 1.]), angle=PI/4 )

        textoda60= MathTex(r"d\vec{A}", font_size=45, color=BLUE).next_to(da60,UP)

        escritaa= MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}| cos(\theta)").move_to(point_or_mobject=[3.5,0,0])
        escrita601= MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}| cos(60^{\circ})").move_to(point_or_mobject=[3.5,0,0])
        escrita602= MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}| (\frac{1}{2})").move_to(point_or_mobject=[3.5,0,0])

        angulo60= Angle(f4, da60, color= PURE_GREEN, radius=0.9)

        GRUPO60= Group(plano2__, da60)

## PARA 90 GRAUS
        plano3__= plano__.copy().rotate(about_point=[-4,0,0], axis=np.array([0., 0., 1.]), angle=PI/2)
        da90= da0.copy().rotate( about_point=[-4,0,0], axis=np.array([0., 0., 1.]), angle=PI/2 )

        textoda90= MathTex(r"d\vec{A}", font_size=45, color=BLUE).next_to(da90,UP+RIGHT)

        escritaaa= MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}| cos(\theta)").move_to(point_or_mobject=[3.5,0,0])
        escrita901= MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}| cos(90^{\circ})").move_to(point_or_mobject=[3.5,0,0])
        escrita902= MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}|(0)").move_to(point_or_mobject=[3.5,0,0])
        escrita903= MathTex(r"\vec{E} \cdot d \vec{A} \, = 0").move_to(point_or_mobject=[3.5,0,0])
        escritanulo= Text(r"Fluxo de campo vetorial nulo", font_size=25).move_to(point_or_mobject=[3.5,-2,0])



        angulo90= Angle(f4, da90, color= PURE_GREEN, radius=0.9)

        tangulo90= MathTex(r"\theta \,", r"= \, 90^{\circ}", font_size=45, color=PURE_GREEN).next_to(angulo90,RIGHT)
        tangulo0= MathTex(r"\theta \,", r" = \, 0^{\circ}", font_size=45, color=PURE_GREEN).next_to(angulo90,RIGHT)
        tangulo60= MathTex(r"\theta \,", r" = \, 60^{\circ}", font_size=45, color= PURE_GREEN).next_to(angulo90,RIGHT)

        GRUPO90= Group(plano3__, da90)
        Grupo900= Group(plano3__, da90)

#FUNCAO VALOR DO COSSENO

        escritacos=MathTex(r"\vec{E} \cdot d \vec{A} \, = \, |d\vec{A}| |\vec{E}|", r"cos(\theta)").move_to(point_or_mobject=[3.5,0,0])
        chavescos= Brace(escritacos[1], DOWN)

        k = ValueTracker(0)
        ax = Axes(x_range=[0, TAU], y_range=[-1, 1])
        graph = ax.plot(lambda x: np.cos(x), color=BLUE, x_range=[0, PI/2])

        valor = always_redraw(lambda: DecimalNumber(num_decimal_places=3)
        .set_value(graph.underlying_function(k.get_value())).next_to(chavescos, 1.5*DOWN)).set_color(PURE_GREEN)

#FUNCAO ANGULO

        rotation_center = [-4,0,0]
        theta_tracker = ValueTracker(0.01)
        da_ref = da.copy()
        planocopia_ref= planofinal.copy()

        planocopia.rotate(
            theta_tracker.get_value()* DEGREES, about_point=rotation_center
         )

        da.rotate(
            theta_tracker.get_value()* DEGREES, about_point=rotation_center
         )


        a = Angle(f4,da, radius=1, other_angle=False, color=PURE_GREEN)

        tangulofinal = MathTex(r"\theta", color=PURE_GREEN).move_to(
            Angle(
                f4, da, radius=1+3 * SMALL_BUFF, other_angle=False
            ).point_from_proportion(0.5)
        )

        planocopia.add_updater(
             lambda x: x.become(planocopia.copy()).rotate(
                theta_tracker.get_value() * DEGREES, about_point=rotation_center
            )
         )

        da.add_updater(
             lambda x: x.become(da.copy()).rotate(
               theta_tracker.get_value() * DEGREES, about_point=rotation_center
            )
         )


        a.add_updater(
             lambda x: x.become(Angle(f4, da, radius=1, color=PURE_GREEN, other_angle=False))
         )

        tangulofinal.add_updater(
            lambda x: x.move_to(
                Angle(
                    f4, da, radius=1+ 3 * SMALL_BUFF, other_angle=False
                ).point_from_proportion(0.5)
            )
        )




#ANIMACOES

        self.play(Write(titulo), run_time=3)
        self.play(GrowArrow(f1_), GrowArrow(f2_), GrowArrow(f3_), GrowArrow(f5_), GrowArrow(f6_), GrowArrow(f7_), run_time=4)
        self.play(ShrinkToCenter(titulo), FadeOut(GRUPOLINHASinicio))

        self.play(FadeIn(GRUPOLINHAS, GRUPO0))

        self.wait(2)

        self.play(FadeIn(textoE, textoda0))
        self.play(Write(escrita))
        self.play(Write(tangulo0))

        self.play(TransformMatchingShapes(escrita, escrita01))
        self.wait(2)
        self.play(TransformMatchingShapes(escrita01, escrita02))
        self.play(Write(escritamax))
        self.wait(2)

        self.play(FadeOut(escritamax), FadeTransformPieces(escrita02, escritaa))


        #self.play(ReplacementTransform(GRUPO0, GRUPO60, run_time=4),

        self.play(Rotate(GRUPO0,angle=PI/4,about_point=[-4,0,0],rate_func=linear, run_time=4),
                  Write(angulo60, run_time=7.1),FadeOut(tangulo0[1]), FadeOut(textoda0))

        self.play(Write(tangulo60[1]), FadeIn(textoda60), run_time=4)
        self.wait(2)

        self.play(TransformMatchingShapes(escritaa, escrita601))
        self.wait(2)
        self.play(TransformMatchingShapes(escrita601, escrita602))
        self.wait(2)
        self.play(FadeTransformPieces(escrita602, escritaaa))

        #self.play(ReplacementTransform(GRUPO60,GRUPO90),
        self.play(Rotate(GRUPO0,angle=PI/4,about_point=[-4,0,0],rate_func=linear, run_time=4), ReplacementTransform(angulo60, angulo90,run_time=4.5),
                  FadeOut(tangulo60[1]), FadeOut(textoda60))


        self.play(Write(tangulo90[1]), FadeIn(textoda90), run_time=4)
        self.wait(2)

        self.play(TransformMatchingShapes(escritaaa, escrita901))
        self.wait(2)
        self.play(TransformMatchingShapes(escrita901, escrita902))
        self.wait(2)
        self.play(TransformMatchingShapes(escrita902, escrita903))
        self.play(Write(escritanulo))
        self.wait(2)

        self.play(FadeOut(GRUPO0, angulo90, tangulo90, tangulo0, textoda90), FadeIn(a), FadeTransformPieces(escrita903, escritacos), FadeOut(escritanulo))
        self.add(da,planocopia, tangulofinal)
        self.wait()

        self.play(Write(valor), Write(chavescos))
        self.wait()

        self.play(theta_tracker.animate.set_value(1.85), k.animate.set_value(PI/2), tangulofinal.animate.set_color(PURE_GREEN), Write(textoda90),
                  run_time=5,
                  rate_func=rate_functions.ease_out_sine)

        self.wait(2)


```

```python
%%manim -ql p3d

class p3d(ThreeDScene):
    def construct(self):
        import numpy as np
        prism = Prism(dimensions=[4, 0.01, 5])
        prism2 = Prism(dimensions=[4, 0.01, 5]).rotate_about_origin(axis=np.array([1., 0., 0.]), angle=PI/4 )
        prism3 = Prism(dimensions=[4, 0.01, 5]).rotate_about_origin(axis=np.array([1., 0., 0.]), angle=PI/2 )
        prism4 = Prism(dimensions=[4, 0.01, 5])

        Grupoflechas1 = VGroup()

        for i in [-2,-1,1,2]:
          Grupoflechas1 += Arrow3D([-2,-3, i],[-2,3,i],
                                thickness=0.01, base_radius= 0.05)

        Grupoflechas2 = VGroup()
        for i in [-2,-1,1,2]:
            Grupoflechas2 += Arrow3D([2,-3, i],[2,3,i],
                                thickness=0.01, base_radius= 0.05)

        Grupoflechas3 = VGroup()
        for i in [-2,-1,1,2]:
            Grupoflechas3 += Arrow3D([0,-3, i],[0,3,i],
                                thickness=0.01, base_radius= 0.05)

        flecha= Arrow3D([0,-3, -0.6],[0,3,-0.6],
                        thickness=0.01, base_radius= 0.05)





        da= Arrow3D(start=np.array([0, 0., 0.]), end=np.array([0, 2., 0.]),color=YELLOW, thickness=0.01, height=0.3, base_radius=0.08)
        da2= Arrow3D(start=np.array([0, 0., 0.]), end=np.array([0, 2., 0.]),color=YELLOW, thickness=0.01, height=0.3, base_radius=0.08
                    ).rotate_about_origin(axis=np.array([1., 0., 0.]), angle=PI/4 )
        da3= Arrow3D(start=np.array([0, 0., 0.]), end=np.array([0, 2., 0.]),color=YELLOW, thickness=0.01, height=0.3, base_radius=0.08
                    ).rotate_about_origin(axis=np.array([1., 0., 0.]), angle=PI/2)
        axes = ThreeDAxes()

        self.set_camera_orientation(phi=75 * DEGREES, theta=30* DEGREES)
        self.add(prism,da, Grupoflechas1,Grupoflechas2, Grupoflechas3)
        self.add(flecha)
        self.wait(2)
        self.play(Transform(prism,prism2), Transform(da,da2))
        self.wait(2)
        self.play(Transform(prism,prism3), Transform(da,da3))
        self.wait(2)
```
